```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

#retrieving current working directory then commenting out
#print(f"Current Working Directory:{Path.cwd()}")
#result:Current Working Directory:/Users/Donya/Desktop/python-challenge/PyRamen

#Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('../PyRamen/menu_data.csv')
sales_filepath = Path('../PyRamen/sales_data.csv')

#Initialize list objects to hold our menu and sales data
menu = []
sales = []

# Initialize a row counter variable
row_count = 0

#Read in the menu data into the menu list
with open(menu_filepath, 'r') as menuCSVFile:
    
    # Determine the data type of the menu_file object
    menu_file_type = type(menuCSVFile)
    print(type(menuCSVFile))
    
    # Create a CSV reader object for the menu data using ',' as the delimiter
    menu_reader = csv.reader(menuCSVFile, delimiter=',')
    print(type(menu_reader))
   
    # Determine the data type of the menu_reader object
    menu_reader_type = type(menu_reader)
    
    # Read and store the header row of the menu data
    menu_header = next(menu_reader)
    
    # Increment the row count
    row_count += 1
    print(f"{menu_header} = MENU HEADER")
    
    # Loop through each line of the menu data and append it to the 'menu' list
    for menu_item in menu_reader:
        menu.append(menu_item)

#Read in the sales data into the sales list
with open(sales_filepath, 'r') as salesCSVFile:
    
    # Determine the data type of the sales_file object
    sales_file_type = type(salesCSVFile)
    print(type(salesCSVFile))
    
    # Create a CSV reader object for the sales data using ',' as the delimiter
    sales_reader = csv.reader(salesCSVFile, delimiter=',')
    print(type(sales_reader))
   
    # Determine the data type of the sales_reader object
    sales_reader_type = type(sales_reader)
    
    # Read and store the header row of the sales data
    sales_header = next(sales_reader)
    
    # Increment the row count
    row_count += 1
    
    # Print the header row
    print(f"{sales_header} = SALES HEADER")
    
    # Loop through each line of the CSV file and append it to the 'sales' list
    for sale_item in sales_reader:
        sales.append(sale_item)

#Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Loop through each sale record in the 'sales' data
for sale_record in sales:
    # Extract data from the sale record
    line_item_id, date, credit_card_number, quantity, sales_item = sale_record

    # Convert 'quantity' to an integer value
    quantity = int(quantity)

    # Check if the 'sales_item' is not already in the 'report' dictionary
    if sales_item not in report:
        # If not, initialize metrics for the new sales item
        report[sales_item] = {
            "01-count": 0,
            "02-revenue": 0,
            "03-cogs": 0,
            "04-profit": 0,
        }

    # Loop through each menu item in the 'menu' data
    for menu_item_info in menu:
        # Extract data from the menu item record
        item, category, description, price, cost = menu_item_info

        # Convert 'price' and 'cost' to float values
        price = float(price)
        cost = float(cost)

        # Calculate profit for the sales item
        profit = price - cost

        # Check if the 'sales_item' matches the current 'item'
        if sales_item == item:
            # If a match is found, update the metrics for the 'sales_item'
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += price * quantity
            report[sales_item]["03-cogs"] += cost * quantity
            report[sales_item]["04-profit"] += profit * quantity
        #else: this works but it made my computer very slow so I am commenting it out
            # If no match is found, print a message
            #print(f"{sales_item} does not equal {item}! NO MATCH!")

    # Increment the row counter by 1
    row_count += 1

# Write the report to a text file
output_path = Path('ramen_report.txt')
with open(output_path, 'w') as file:
    file.write('Ichiban Ramen Sales Report\n')
    for sales_item, metrics in report.items():
        file.write(f"{sales_item} {metrics}\n")
```
